```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
from keras.models import load_model
import numpy as np
from PIL import Image


# Load the model
amodel = load_model('harmodel.h5')

def read_img(img_path):
    img = Image.open(img_path)
    return np.asarray(img.resize((150,150)))

# Function to predict and display results
def test_predict(test_image, label_var, image_label):
    result = amodel.predict(np.asarray([read_img(test_image)]))
    print(result.shape)
    # Assuming your CSV processing code looks something like this
    labels=['calling','clapping','cycling','dancing','drinking','eating','fighting','hugging','laughing','listening_to_music','running','sitting','sleeping','texting','using_laptop']

    predicted_class = np.argmax(result)
    probability = np.max(result) * 100

    predicted_label = labels[predicted_class]  # Get the label text from the mapping

    print("Probability: {:.2f}%".format(probability))
    print("Predicted class:", predicted_class)
    print("Predicted label:", predicted_label)

    # Display the image with the predicted class and label text as the title
    image = Image.open(test_image)  # Use PIL to open the image
    image = image.resize((400, 400))  # Resize the image
    photo = ImageTk.PhotoImage(image)

    image_label.configure(image=photo)
    image_label.image = photo

    # Update label with predicted result
    label_var.set("Predicted class: {}, Label: {}".format(predicted_class, predicted_label))

# Create a Tkinter window
root = tk.Tk()

# Set window size
window_width = 800
window_height = 600
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
x_position = (screen_width - window_width) // 2
y_position = (screen_height - window_height) // 2
root.geometry("{}x{}+{}+{}".format(window_width, window_height, x_position, y_position))

# Function to select an image file
def select_image():
    initial_dir = "hartest"
    file_path = filedialog.askopenfilename()
    test_predict(file_path, label_var, image_label)

# Button to trigger the test_predict function
label_var = tk.StringVar()
label_var.set("Prediction")
result_label = tk.Label(root, textvariable=label_var)
result_label.pack()

image_label = tk.Label(root)
image_label.pack()

button = tk.Button(root, text="Select Image", command=select_image)
button.pack()

# Run the Tkinter event loop
root.mainloop()
```
